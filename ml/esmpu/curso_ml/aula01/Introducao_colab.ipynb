### 2- Criar link para leitura dos dados por outros colegas da equipe do projeto

#### Como compartilhar dados com link direto pro Google Drive
Fazer upload dos dados pro seu Google Drive.  
Compartilhar o arquivo (opção: Qualquer pessoa com o link. Leitor).  
Copiar o link e modificar o link conforme abaixo:     

Link padrão p/ compartilhamento:     
"!wget --no-check-certificate 'https://docs.google.com/uc?export=download&id=FILEID' -O FILENAME"   
Onde FILEID é retirado do link original de compartilhamento do arquivo.  


Exemplo: 
df_enem2019_socioecon.csv   
Link original = "https://drive.google.com/file/d/1np3BsL_1SBj_aqO8xfna9YvO0T4R3yFe/view?usp=sharing" .   

Neste caso, o FILEID do link é "1np3BsL_1SBj_aqO8xfna9YvO0T4R3yFe"   

**Portanto, o link final para download será:**    
!wget --no-check-certificate 'https://docs.google.com/uc?export=download&id=1np3BsL_1SBj_aqO8xfna9YvO0T4R3yFe' -O df_enem2019_socioecon.csv   


Agora crie um link compartilhável pro seu arquivo e tente ler diretamente.
