Make Suitable Plots of the follwing Features num-of-doors body-style drive-wheels engine-location wheel-base stroke fuel-type horsepower price

```python
sns.set_theme(style="darkgrid")

ax = sns.countplot(x="num-of-doors", data= df_car)
#how many different num of doors with count value
```

```python
ax = sns.countplot(x="body-style", data= df_car)
#how many different body styles with count
```

```python
sns.catplot(x="drive-wheels", y="wheel-base", data= df_car)
#cat plot for driver wheel and base wheel
```

```python
sns.countplot(df_car['engine-location']);
```

```python
sns.kdeplot(df_car['stroke']);
#density plot for stroke
```

```python
sns.catplot(x="fuel-type", y="horsepower", data=df_car )
#effect on horsepower due to fuel type
```

```python
sns.catplot(x="fuel-type", y="price", hue="engine-type", kind="swarm", data= df_car)
#plotting impact of fuel type on price and engine type
```

```python
corr = df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
#correlation between df columns
```
