```python
# Importação das bibliotecas
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv("../dados/co2.csv")
base.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
srn.scatterplot(data=base, x=base.conc, y=base.uptake, hue=base.Type)
```

```python
# Seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base.loc[base["Type"] == "Quebec"]
m = base.loc[base["Type"] == "Mississippi"]
```

```python
# Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(data=q, x=q.conc, y=q.uptake).set_title("Quebec")
plt.subplot(1, 2, 2)
srn.scatterplot(data=q, x=q.conc, y=q.uptake).set_title("Mississippi")
plt.tight_layout()
```

```python
# Refrigerado e não refrigerado
ch = base.loc[base["Treatment"] == "chilled"]
nc = base.loc[base["Treatment"] == "nonchilled"]
```

```python
# Gráfico somente com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1, 2, 1)
srn.scatterplot(data=ch, x=ch.conc, y=ch.uptake).set_title("Chilled")
plt.subplot(1, 2, 2)
srn.scatterplot(data=nc, x=nc.conc, y=nc.uptake).set_title("Non chilled")
plt.tight_layout()
```

```python
# Carregamento de outro arquivo, cancer de esofago
base2 = pd.read_csv("../dados/esoph.csv")
base2
```

```python
# cor padrao da srn
colorpalette = srn.color_palette()

# Gráfico entre os atributos 'alcgp' e 'ncontrols'
srn.catplot(x="alcgp", y="ncontrols", data=base2, jitter=False, palette=colorpalette)
```

```python
# Gráfico entre os atributos 'alcgp' e 'ncontrols', com agrupamento
# pcolor = srn.color_palette("Spectral")
srn.catplot(
    x="alcgp", y="ncontrols", data=base2, col="tobgp", palette=colorpalette
)  # palette = ['blue', 'red', 'green', 'yellow'])
```
